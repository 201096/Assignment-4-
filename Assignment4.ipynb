```python
# команда для вывода всех сотрудников у которых есть пациенты


query = '''SELECT employees.employee_id,employee_data.full_name,employee_data.post,COUNT(patient_id) AS amount
FROM employees
INNER JOIN employee_data ON employee_data.employee_id=employees.employee_id
INNER JOIN treatment ON employees.employee_id=treatment.employee_id
GROUP BY employees.employee_id,employee_data.full_name,employee_data.post;'''
data = pd.read_sql_query(query, conn)
data
```

```python
# команда вывода пациента на лечении 
query = '''SELECT patients.full_name,departments.name_d,treatment.diagnosis,schedule_treatment.start_of_treatment
FROM patients
INNER JOIN treatment ON patients.patient_id=treatment.patient_id
INNER JOIN employees ON treatment.employee_id=employees.employee_id
INNER JOIN departments ON employees.department_id=departments.department_id
INNER JOIN schedule_treatment ON treatment.treatment_id=schedule_treatment.treatment_id
WHERE patients.patient_id=503;'''
data = pd.read_sql_query(query, conn)
data
```

```python
# команда для вывода инфо для пациента 

query = '''SELECT patients.patient_id,patients.full_name,patients.ward,patients.special_group,medical_history.birth_wegth,medical_history.growth_at_birth,medical_history.vaccinations,medical_history.full_medical_document
FROM patients
INNER JOIN medical_history ON medical_history.patient_id=patients.patient_id
WHERE patients.patient_id = 503;'''
data = pd.read_sql_query(query, conn)
data
```

```python
# запрос для вывода среднего опыта работы по определенному отделу

query = '''SELECT ROUND(AVG(salary),2) AS avgsalary
FROM employee_data
INNER JOIN employees ON employees.employee_id=employee_data.employee_id
'''
data = pd.read_sql_query(query, conn)
data
```

```python
# запрос для нахождения максимлаьной зарплаты в определенном отделе

query = '''SELECT employee_data.full_name,employee_data.post,employee_data.salary
FROM employee_data
INNER JOIN employees ON employees.employee_id=employee_data.employee_id
WHERE employee_data.salary IN
(SELECT MAX(salary) FROM employee_data INNER JOIN employees ON employees.employee_id=employee_data.employee_id);'''
data = pd.read_sql_query(query, conn)
data
```
